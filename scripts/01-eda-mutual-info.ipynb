```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
def load_files(background='../data/FFChallenge_v5/background.csv', train='../data/FFChallenge_v5/train.csv'):

    background = pd.read_csv(background, low_memory=False)  # ,\

    train = pd.read_csv(train, sep=',', header=0,
                        index_col=0).dropna(how='all')
    data = background.loc[train.index].join(train)
    
    return data
```

```python
img = sns.heatmap(miscores)

img.set_yticklabels([])
img.set_yticks([])

img.xaxis.tick_top()

img.set_xticklabels(img.get_xticklabels(), rotation=45)

# add label mutual information to the scale
img.figure.axes[-1].set_ylabel('Mutual information with independent variables', rotation=270, labelpad=20)

img.figure.set_size_inches(8, 8)

plt.show()
img.figure.savefig('../images/01-miscores.png')
```
